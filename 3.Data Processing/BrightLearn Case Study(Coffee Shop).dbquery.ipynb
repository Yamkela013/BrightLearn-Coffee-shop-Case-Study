```sql
  SELECT *
FROM bright_coffee_shop_sales
LIMIT 10;
----------------
---CHECK DATE----------
--------------
--they started collecting the data 2023-01-01
--------------------

SELECT MIN(transaction_date) AS min_date
FROM bright_coffee_shop_sales;
-----------------------
--The duration of the data is 6 months
--Last collected data on 2023-06-30
-----------------------------

SELECT MAX(transaction_date) AS latest_date
FROM bright_coffee_shop_sales;

------------------------------------

---CHECKING NAMES OF DIFFERENT STORES
  --We have 3 stores, Lower Manhattan, Hell's Kitchen,Astoria
------------------------------------

SELECT DISTINCT store_location
FROM bright_coffee_shop_sales;

----------------------------------------
---CHECKING PRODUCTS SOLD at our stores
----------------------------------------
SELECT DISTINCT product_category
FROM bright_coffee_shop_sales;
----------------------------------
----------------------------------------------------------
---DATA PROCESSING ----
SELECT
------DATES---
       transaction_date AS purchase_date,
       dayname(transaction_date) AS Day_name,
       Monthname(transaction_date) AS Month_name,
      date_format(transaction_time, 'HH:mm:ss') AS purchase_time,
      dayofmonth(transaction_date) AS day_of_month,
   CASE 
     WHEN dayname(transaction_date) IN ('Sun' , 'Sat') THEN 'Weekend'
     ELSE 'Weekday'
  END AS day_classification,

  CASE  
       WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '00:00:00' AND '11:59:59' THEN '01. Morning'
      WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '12:00:00' AND '17:59:59' THEN '02.Afternoon'
      WHEN date_format(transaction_time, 'HH:mm:ss') >= '17:00:00' THEN 'Evening'
  END AS time_buckets,

---COUNTS OF ID'S----
       COUNT(DISTINCT transaction_id) AS Number_of_Sales,
       COUNT(DISTINCT product_id) AS number_of_products,
       COUNT(DISTINCT store_id) AS number_of_stores,
----REVENUE BY DAY----
 SUM(transaction_qty*unit_price) AS revenue_per_day,
  
  CASE 
     WHEN revenue_per_day <= 50 THEN '01.Low Spend'
     WHEN revenue_per_day BETWEEN 51 AND 100 THEN '02.Medium Spend'
     ELSE '03.High Spend'
  END AS spend_buckets,

 ----CATEGORICAL COLUMNS-
    product_category,
     store_location,
     product_detail
FROM bright_coffee_shop_sales
GROUP BY purchase_date,
         Day_name,
         Month_name,
         store_location,
        product_category,
         product_detail,
         purchase_time,
         time_buckets,
         day_classification;
----END---------
------------------------------------
----NB!!! We group by the columns that are not aggreagated within our Select Statement
------------------------------------



```
